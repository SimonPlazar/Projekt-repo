```python
import cv2 as cv
import os
import numpy as np
import dlib
```

```python
index = 7
img_dir_name = f'raw_data{index}'
label_dir_name = f'label{index}'

par_dir = 'data'
path_img = os.path.join(par_dir, img_dir_name)
path_label = os.path.join(par_dir, label_dir_name)

if not os.path.exists(path_img):
    os.makedirs(path_img)
    print("Img dir ustvarjen")
else:
    print("img dir ze obstaja!")

if not os.path.exists(path_label):
    os.makedirs(path_label)
    print("Label dir ustvarjen")
else:
    print("Label dir ze obstaja!")
```

```python
def save_data(frame,angle,zaprte):
    idx = len(os.listdir(f'data/{img_dir_name}'))
    print(f"Stevilo slik: {idx}")
    frame_fd = f'data/{img_dir_name}/IMG{idx}.png'
    label_fd = f'data/{label_dir_name}/Label{idx}.txt'

    cv.imwrite(frame_fd, frame)
    with open(label_fd, 'w') as f:
        f.write(f'{angle} {zaprte}')
    
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('models/shape_predictor_68_face_landmarks.dat')

capture = cv.VideoCapture(1)

while True:
    ret, frame = capture.read()
    frame_train = frame.copy()
    if not ret:
        break

    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)


    faces = detector(gray)

    for face in faces:
        landmarks = predictor(gray, face)

        left_eye_center = ((landmarks.part(36).x + landmarks.part(39).x) // 2,
                           (landmarks.part(36).y + landmarks.part(39).y) // 2)
        right_eye_center = ((landmarks.part(42).x + landmarks.part(45).x) // 2,
                            (landmarks.part(42).y + landmarks.part(45).y) // 2)

        delta_x = right_eye_center[0] - left_eye_center[0]
        delta_y = right_eye_center[1] - left_eye_center[1]

        angle = np.arctan(delta_y / delta_x)

        angle = (angle * 180) / np.pi

        cv.circle(frame, left_eye_center, 2, (0, 255, 0), -1)
        cv.circle(frame, right_eye_center, 2, (0, 255, 0), -1)

        cv.imshow('Frame', frame)
        print(f"Shranim sliko? Kot je: {angle} stopinj.")
        key = cv.waitKey(3000) & 0xFF
        if key == 27:
            break
        elif key == 32:
            while True:
                print(f"Določi zaprtost uči, I za zaprte 0 za odprte")
                zaprto = cv.waitKey(0) & 0xFF
                if zaprto == ord("i"):
                    zaprto = 1
                    break
                elif zaprto == ord("o"):
                    zaprto = 0
                    break
            save_data(frame_train, angle, zaprto)
    if cv.waitKey(1) & 0xFF == 27:
        break

capture.release()
cv.destroyAllWindows()
```
